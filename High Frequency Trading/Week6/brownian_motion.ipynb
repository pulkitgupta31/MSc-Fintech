```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate Brownian motion
def brownian_motion(num_steps, delta_t, delta):
    """
    Simulate a Brownian motion
    :param num_steps: Number of steps
    :param delta_t: Time interval
    :param delta: Scale factor for the normal distribution (sqrt of variance)
    :return: A numpy array representing the Brownian motion
    """
    # Random steps drawn from normal distribution
    steps = np.random.normal(loc=0, scale=np.sqrt(delta_t) * delta, size=num_steps)
    # Cumulative sum to simulate the motion
    brownian_motion = np.cumsum(steps)
    # Insert the initial condition (starting at 0)
    brownian_motion = np.insert(brownian_motion, 0, 0)
    return brownian_motion

# Parameters
num_steps = 1000  # Number of steps in the Brownian motion
delta_t = 1       # Time interval
delta = 1         # Scale factor for step size

# Simulate the Brownian motion
bm = brownian_motion(num_steps, delta_t, delta)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(bm)
plt.title('Brownian Motion')
plt.xlabel('Time')
plt.ylabel('Position')
plt.grid(True)
plt.show()
```
